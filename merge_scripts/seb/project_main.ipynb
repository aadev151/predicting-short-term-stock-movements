```python
import pandas as pd
```

```python
sp = pd.read_csv('base_data/sp500_simple.csv', parse_dates=['date']).convert_dtypes()
sp.head()
```

```python
comp = pd.read_csv('gmm_regimes.csv', parse_dates=['Date']).convert_dtypes()
comp.Regime_GMM = comp.Regime_GMM.astype('category')
comp.Regime_label = comp.Regime_label.astype('category')
comp = comp[comp.Date.between(sp.date.min(), sp.date.max())].reset_index(drop=True)
comp.head()
```

```python
merged = pd.merge(sp, comp, left_on=['date'], right_on=['Date'])
merged.drop(inplace=True, columns=['SP500_Return', 'Date'])
merged.head()

merged.to_csv('seb.csv', index=False)
```
